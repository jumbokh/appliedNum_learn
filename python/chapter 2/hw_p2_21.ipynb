```python
from newtonx import *
```

```python
def f(x):
    return -5*x**4+11*x**2-2
```

```python
def fpr(x):
    return -20*x**3+22*x
```

```python
x,logdat = newtonx(f,fpr,1.0,10,tol = 1.0e-4)
#print('x = ','{:8.6f}'.format(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(-1.0,3.0,.01)
n = np.copy(t)
n.fill(0)
plt.plot(t, f(t), 'r')
plt.plot(t,n,'k')
for a,fa,m,fm in logdat:
    plt.plot(a,fa,'go')
    plt.plot(m,fm,'bo')
plt.show()
```

```python
def fpr1(x):
    return 20*x**3-22*x
```
