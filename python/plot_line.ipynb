```python
from secantx import *
from prettytable import PrettyTable
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**5-0.5
```

```python
x,logdat = secantx(f,0,1.0,10,tol = 1.0e-4)

printx = PrettyTable(["a","b","m","ym" ])
printx.padding_width = 1
    
for a,b,m,fm in logdat:
    printx.add_row([a,b,m,fm])   
print(printx)
```

```python
t = np.arange(-1.5,1.2,.01)
n = np.copy(t)
n.fill(0)
plt.plot(t, f(t), 'r')
plt.plot(t,n,'k')
for a,b,m,fm in logdat:
    #plt.plot(a,f1,color = 'g')
    plt.scatter(a,f(a),color='b')
    #plt.plot(b,f2,'bo')
    plt.scatter(b,f(b),color='y')
    #plt.plot(m,f3,color='c')
    plt.scatter(m,fm,color='c')
plt.show()
```

```python
t = np.arange(-1.5,1.2,.001)
n = np.copy(t)
n.fill(0)
plt.plot(t, f(t), 'r')
plt.plot(t,n,'k')
for a,b,m,fm in logdat:
    #plt.plot(a,f1,color = 'g')
    plt.scatter(a,f(a),color='b')
    #plt.plot(b,f2,'bo')
    plt.scatter(b,f(b),color='y')
    #plt.plot(m,f3,color='c')
    plt.scatter(m,fm,color='c')
x = []
y = []
for a,b,m,fm in logdat:
    x.append([a,f(a)])
    y.append([b,f(b)])
for i in range(len(x)):
    plt.plot(x[i], y[i], color='y')
    plt.scatter(x[i], y[i], color='g')    
```
