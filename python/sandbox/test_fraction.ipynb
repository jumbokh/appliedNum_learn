```python
size = 3

# We want to invert the following matrix:

toInverse = [[0 for j in range(size)] for i in range(size)]
toInverse[0] = [3, -1, 2]
toInverse[1] = [1, 1, 2]
toInverse[2] = [3, -2, -1]
matrix = toInverse.copy()
#print(inverse(toInverse)) # "None" because the matrix is not invertible!
final = inverse(toInverse)
print("--------------------------")
pmat(final)
```

```python
import numpy as np
c = np.dot(matrix,final)
pmat(matrix)
print("--------------")
pmat(final)
print("==============")
pmat(c)
```
