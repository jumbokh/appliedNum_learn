```python
from power_B import *
from rayleigh import *
from Aitken import *
import numpy as np
```

```python
A = np.array([[40.5,10.5,-21.0,21.0],[10.5, 20.5, -5.0, 5.0],[-21.0, -5.0, 37.0, -1.0],[21.0, 5.0, -1.0, 37.0]])
z = [[1.0],[1.0],[1.0],[1.0]]
print(A)
```

```python
np.set_printoptions(precision=5)
[eigs,vecs] = power_B(A,z,16,0.0001)
print('eigs: ',eigs)
for x in vecs.T:
    print(x,end='')
print()
```

#### 將矩陣H平移20  (H = A - 20*I)

```python
H = A - 20 * np.eye(4)
print(H)
```

```python
[eigs,vecs] = power_B(H,z,15,0.0001)
print('eigs: ',eigs)
for x in vecs.T:
    print(x,end='')
print()
```

```python
[eigs,vecs] = rayleigh(H,z,15,0.0001)
print('eigs: ',eigs)
for x in vecs.T:
    print(x,end='')
print()
```

```python
[eigs,vecs,Ait] = Aitken(H,z,15,0.0001)

for x in vecs.T:
    print(x,end='')
print()
print('Power_RQ:')
for t in eigs:
    print('%8.4f'% t,', ',end='')
    if t==5 :
        print()
print()
print('Aitken:')
for y in Ait:
    print('%8.4f'% y,', ',end='')
    if y==5:
        print()
print()
```
